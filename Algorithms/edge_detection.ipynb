```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../data/images/source_color/source_image_3_color.jpg', cv.IMREAD_GRAYSCALE)

_, thresh = cv.threshold(img, 0, 255, cv.THRESH_BINARY_INV + cv.THRESH_OTSU)

kernel = np.ones((3, 3), np.uint8)

grad = cv.morphologyEx(thresh, cv.MORPH_GRADIENT, kernel)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(grad, cmap='gray')
plt.axis('off')

plt.show()
```
